### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.ettoday.net/news/news-list.htm'
web = requests.get(url)
soup = BeautifulSoup(web.text, exclude_encodings="lxml")
newList = soup.find(class_='part_list_2').findAll('h3')
```

```python
for i in newList:
    print(i.find(class_='date').text, i.find('em').text, i.find('a').text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
dataList = {}
for i in newList:
#     print(dataList.get(i.find('em').text) is None)
    if dataList.get(i.find('em').text) is None:
        dataList[i.find('em').text] = [i.find('a').text]
    else :
        dataList[i.find('em').text].append(i.find('a').text)
        
dataList
```

### ③ 哪一個類別的文章最多

```python
maxType = [];
maxNum = 0
for i in dataList.keys():
    if len(dataList[i]) > maxNum :
        maxType = [i]
        maxNum = len(dataList[i]); 
    elif len(dataList[i]) == maxNum :
        maxType.append(i)
        maxNum = len(dataList[i]);

    
print('最多篇數的分類%s:%d' % (maxType, maxNum))    
        
        
        
```
