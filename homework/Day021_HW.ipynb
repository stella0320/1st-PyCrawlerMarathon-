## 1. 取出今天所有的新聞

```python
import requests
import datetime
import time
dateNow = datetime.datetime.today()
today = datetime.datetime(dateNow.year, dateNow.month, dateNow.day)
url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
soup = BeautifulSoup(r.text, "html5lib")

finalDateStr = soup.find(class_="part_list_2").find_all('h3')[-1].find('span', attrs={'class', 'date'}).text
finalDate = datetime.datetime.strptime(finalDateStr, '%Y/%m/%d %H:%M')
while(finalDate > today):
    browser.execute_script("window.scrollTo(0, 10000);")
    time.sleep(2)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    finalDateStr = soup.find(class_="part_list_2").find_all('h3')[-1].find('span', attrs={'class', 'date'}).text
    finalDate = datetime.datetime.strptime(finalDateStr, '%Y/%m/%d %H:%M')
```

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    newsTime = d.find(class_="date").text
    newsTime = datetime.datetime.strptime(newsTime, '%Y/%m/%d %H:%M')
    if (newsTime > today):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
towHour = datetime.datetime(dateNow.year, dateNow.month, dateNow.day, dateNow.hour-2)
for d in soup.find(class_="part_list_2").find_all('h3'):
    newsTime = d.find(class_="date").text
    newsTime = datetime.datetime.strptime(newsTime, '%Y/%m/%d %H:%M')
    if (newsTime > towHour):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
import requests
import datetime
import time
dateNow = datetime.datetime.today()
today = datetime.datetime(dateNow.year, dateNow.month, dateNow.day)
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")
startTime = datetime.datetime(dateNow.year, dateNow.month, dateNow.day-3, 15)
endTime = datetime.datetime(dateNow.year, dateNow.month, dateNow.day-3, 17)
newsTime = dateNow
while(startTime < newsTime):
    print('-----')
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    finalTime = datetime.datetime.strptime(soup.find(class_="part_list_2").find_all('h3')[-1].find(class_="date").text, '%Y/%m/%d %H:%M')
    print('final time:%s, start time:%s, finalTime < startTime:%s'% (finalTime, startTime, finalTime < startTime))
    if (finalTime < startTime):
        for d in soup.find(class_="part_list_2").find_all('h3'):
            newsTimeStr = d.find(class_="date").text
            newsTime = datetime.datetime.strptime(newsTimeStr, '%Y/%m/%d %H:%M')
            title = d.find_all('a')[-1].text
            if (newsTime >= startTime and endTime >= newsTime):
                print(newsTimeStr, title)
    else:
        browser.execute_script("window.scrollTo(0, 10000);")
        time.sleep(5)
```
