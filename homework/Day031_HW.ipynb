### 尋找一個網站是有作 Header 檢查的（Hint: 有加跟沒有加回傳結果不一樣）

```python
import requests
url = 'https://site.douban.com/214956/widget/notes/13925727/note/381435207/'
response = requests.get(url)

response.text
```

### 用上述網站說明該如何判斷 Header 中必須加上哪些資料才會正確


```python
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.100 Safari/537.36'}

response = requests.get(url, headers = header)
response.text
```
